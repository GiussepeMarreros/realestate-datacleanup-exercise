#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

```python
# TODO
from sklearn.preprocessing import MinMaxScaler

cinturon_sur_pob= cinturon_sur_pob.copy()

norm_df=MinMaxScaler()
cinturon_sur_pob['precio_minmax']= norm_df.fit_transform(cinturon_sur_pob[['price']])

poblaciones_norm= cinturon_sur_pob['level5'].unique()
plt.figure(figsize=(12,8))

for i, poblacion in enumerate(poblaciones_norm):
    plt.subplot(2,2,i+1)
    data_poblacion= cinturon_sur_pob[cinturon_sur_pob['level5']==poblacion]
    plt.hist(data_poblacion['precio_minmax'])
    plt.title(f' Precio normalizado en {poblacion}')
    plt.xlabel('Precio normalizado')
    plt.ylabel('Frecuencia')


plt.tight_layout()
plt.show

#cinturon_sur_pob['precio_minmax']
```
